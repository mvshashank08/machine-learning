```python
import matplotlib
import matplotlib.pyplot as plt

pos = 69000
some_digit = X[pos]
digit_image = some_digit.reshape(28, 28)
#digit_image

plt.imshow(digit_image, cmap = matplotlib.cm.binary, interpolation="nearest")
#plt.axis("off")
plt.show()
y[pos]
```
